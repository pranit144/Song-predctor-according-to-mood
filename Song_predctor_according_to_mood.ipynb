```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import joblib

# -------------------------------
# 1. Load your dataset
# -------------------------------
data_path = 'data_moods.csv'  # Path to your CSV file
df = pd.read_csv(data_path)

# -------------------------------
# 2. Encode categorical features
# -------------------------------
# Encode 'artist'
le_artist = LabelEncoder()
df['artist_encoded'] = le_artist.fit_transform(df['artist'])

# Encode 'mood'
le_mood = LabelEncoder()
df['mood_encoded'] = le_mood.fit_transform(df['mood'])

# -------------------------------
# 3. Prepare input features and target
# -------------------------------
# Our input features are the encoded artist and mood
X = df[['artist_encoded', 'mood_encoded']]

# Define the target columns (parameters your model will predict)
target_columns = [
    'danceability', 'acousticness', 'energy', 'instrumentalness',
    'liveness', 'valence', 'loudness', 'speechiness', 'tempo', 'key',
    'time_signature'
]
y = df[target_columns]

# -------------------------------
# 4. Scale the target parameters
# -------------------------------
# Scaling is important when targets have different scales.
scaler = StandardScaler()
y_scaled = scaler.fit_transform(y)

# -------------------------------
# 5. (Optional) Train/Test Split
# -------------------------------
# You can perform a train-test split to evaluate your model.
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.2, random_state=42)

# -------------------------------
# 6. Train the Model
# -------------------------------
# Using MultiOutputRegressor to handle multiple target outputs
base_model = RandomForestRegressor(n_estimators=100, random_state=42)
model = MultiOutputRegressor(base_model)
model.fit(X_train, y_train)

# (Optional) Evaluate the model performance on the test set
# from sklearn.metrics import mean_squared_error
# y_pred = model.predict(X_test)
# mse = mean_squared_error(y_test, y_pred)
# print("Mean Squared Error on Test Data:", mse)

# -------------------------------
# 7. Save the model and supporting objects
# -------------------------------
joblib.dump(model, 'optimized_random_forest_model.pkl')
joblib.dump(le_artist, 'artist_encoder.pkl')
joblib.dump(le_mood, 'mood_encoder.pkl')
joblib.dump(scaler, 'scaler.pkl')

print("Training complete and model saved!")
```

```python
import joblib
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

# -------------------------------
# 1. Load saved artifacts and data
# -------------------------------
model_path = 'optimized_random_forest_model.pkl'
artist_encoder_path = 'artist_encoder.pkl'
mood_encoder_path = 'mood_encoder.pkl'
scaler_path = 'scaler.pkl'
data_path = 'data_moods.csv'

model = joblib.load(model_path)
le_artist = joblib.load(artist_encoder_path)
le_mood = joblib.load(mood_encoder_path)
scaler = joblib.load(scaler_path)
df = pd.read_csv(data_path)

# -------------------------------
# 2. Define sample input (artist and mood)
# -------------------------------
artist_input = "Prince"   # Change as needed; must be present in training data.
mood_input = "Happy"      # Change as needed.

# Encode the input values using the saved label encoders.
artist_encoded = le_artist.transform([artist_input])[0]
mood_encoded = le_mood.transform([mood_input])[0]

# Create a DataFrame with the same feature names as during training.
X_test = pd.DataFrame([[artist_encoded, mood_encoded]], columns=['artist_encoded', 'mood_encoded'])

# -------------------------------
# 3. Predict target features using the model
# -------------------------------
# The model predicts scaled target parameters.
y_scaled_pred = model.predict(X_test)[0]

# Inverse transform to get the original scale of the target parameters.
predicted_features = scaler.inverse_transform([y_scaled_pred])[0]

# Define the target columns (the order should match training).
target_columns = [
    'danceability', 'acousticness', 'energy', 'instrumentalness',
    'liveness', 'valence', 'loudness', 'speechiness', 'tempo', 'key',
    'time_signature'
]

# Create a DataFrame for the predicted features.
predicted_df = pd.DataFrame([predicted_features], columns=target_columns)

# -------------------------------
# 4. Compute distances to each song in the dataset
# -------------------------------
# Select the same target parameters from the dataset.
song_features = df[target_columns]

# Calculate Euclidean distances between each song's parameters and the predicted parameters.
distances = euclidean_distances(song_features, predicted_df)

# Add distances as a new column in the dataset.
df['distance'] = distances

# Sort the songs by increasing distance (closest match first) and pick the top 5.
sorted_df = df.sort_values(by='distance').head(5)

# Optionally, add a link/button for a YouTube search.
sorted_df['YouTube'] = sorted_df['name'].apply(
    lambda song: f"https://www.youtube.com/results?search_query={song}"
)

# -------------------------------
# 5. Output the recommended song list
# -------------------------------
recommended_songs = sorted_df[['name', 'album', 'artist', 'popularity', 'YouTube']]
print("Recommended Songs:")
print(recommended_songs.to_string(index=False))
```
